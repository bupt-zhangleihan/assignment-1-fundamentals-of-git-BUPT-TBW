```python
url = 'https://en.wikipedia.org/wiki/Data_science'
import requests
url = 'https://en.wikipedia.org/wiki/Data_science'
#url = 'http://127.0.0.1:5500/1-Introduction/01-defining-data-science/Data%20science%20-%20Wikipedia.html'
text = requests.get(url,timeout=500).content.decode('utf-8')
print(text[:1000])
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    script = False
    res = ""
    def handle_starttag(self, tag, attrs):
        if tag.lower() in ["script","style"]:
            self.script = True
    def handle_endtag(self, tag):
        if tag.lower() in ["script","style"]:
            self.script = False
    def handle_data(self, data):
        if str.strip(data)=="" or self.script:
            return
        self.res += ' '+data.replace('[ edit ]','')

parser = MyHTMLParser()
parser.feed(text)
text = parser.res
print(text[:1000])
import sys
#配置阿里云pypi安装源
!{sys.executable} -m pip config set global.index-url https://mirrors.aliyun.com/pypi/simple
import sys
!{sys.executable} -m pip install nlp_rake
import nlp_rake
import os
os.makedirs('images', exist_ok=True) # 创建 images 文件夹

extractor = nlp_rake.Rake(max_words=2,min_freq=3,min_chars=5)
res = extractor.apply(text)
res
import matplotlib.pyplot as plt

plt.gcf().subplots_adjust(left=0.05,top=0.91,bottom=0.09) 

plt.rcParams['font.family'] = ['Arial', 'Calibri', 'Times New Roman']#找不到Helvetica字体，换成Calibri后成功运行
plt.rcParams['axes.unicode_minus'] = False

def plot(pair_list):
    plt.figure(figsize=(12,6)).patch.set_facecolor('white')
    k,v = zip(*pair_list)
    #print(k,v)
    plt.bar(range(len(k)),v)
    plt.xticks(range(len(k)),k,rotation='vertical',fontsize=16)
    plt.tight_layout()
    plt.savefig('images/bar.png',dpi=600)
    plt.show()

plot(res[:20])
!{sys.executable} -m pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc = WordCloud(background_color='white',width=800,height=600)
plt.figure(figsize=(15,7)).patch.set_facecolor('white')
plt.imshow(wc.generate_from_frequencies({ k:v for k,v in res }))
plt.figure(figsize=(15,7)).patch.set_facecolor('white')
plt.imshow(wc.generate(text))
wc.generate(text).to_file('images/ds_wordcloud.png')
print("思考科学数据场景")
print("收集社交媒体上的文本数据，如推特或微博上的评论和帖子。如果您想了解某个地区的气候变化，您可以收集气象数据，如温度、降雨量、风速等。收集数据的方法也取决于数据的类型和来源。有些数据可以通过公开的网站或数据库获得，如政府统计数据、新闻报道、学术论文等。有些数据需要使用特定的工具或技术来获取，如网络爬虫、传感器、问卷调查等。有些数据需要经过人工筛选或标注，如图像、视频、音频等。存储数据的方式也取决于数据的类型和规模。有些数据可以存储在本地计算机或移动设备上，如文本文件、电子表格、图片等。有些数据需要存储在云端或分布式系统上，如大型数据库、数据仓库、数据湖等。数据的大小可能因不同的因素而异，如采样频率、分辨率、编码格式等。一般来说，越多维度和特征的数据越大，越多样本和观察值的数据越大。从数据中获得见解的方法也取决于数据的类型和目标。有些数据可以通过简单的描述性统计或可视化来分析，如平均值、标准差、频率分布、条形图、折线图等。有些数据需要使用复杂的推理性统计或机器学习来分析，如假设检验、回归分析、聚类分析、分类分析等。从数据中获得的见解可以帮助我们理解现象、发现规律、预测结果、优化策略等。根据数据做出决定时，需要考虑数据的质量、有效性、可靠性和代表性等因素。")
```
